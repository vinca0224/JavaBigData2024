- 챗GPT용 함수 생성

````python
def query_gpt35(text):
    res = client.chat.completions.create(
        model = 'gpt-3.5-turbo',
        messages = [
                        # {'role' : 'system', 'content' : 'You are a helpful assistant.'},
                        {'role' : 'user', 'content' : text},
                        # {'role' : 'assistant', 'content' : " Who's there?"}
                    ]
        )
    return res.choices[0].message.content
````
